```python
import pandas as pd

df = pd.read_json('yelp_academic_dataset_review.json',lines=True)
```

```python
df
```

```python
Positive_stars = df[(df['stars'] == 4) | (df['stars'] == 5)].sample(25000)
```

```python
Positive_stars
```

```python
Neutral_stars = df[(df['stars'] == 3)].sample(25000)
```

```python
Neutral_stars
```

```python
Negative_stars = df[(df['stars'] == 1) | (df['stars'] == 2)].sample(25000)
```

```python
Negative_stars
```

```python
df_List = [Positive_stars, Neutral_stars, Negative_stars]
New_df = pd.concat(df_List)
```

```python
New_df
```

```python
shuffled_Df = New_df.sample(frac=1)
```

```python
shuffled_Df
```

```python
shuffled_Df.loc[shuffled_Df['stars'] == 2]
```

```python
shuffled_Df.to_csv(path_or_buf="Shuffled_CSV.csv", index=False)
```
